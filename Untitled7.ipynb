## Formulate Hypothesis

### Subtask:
Clearly define the null and alternative hypotheses for the A/B test regarding the impact of the new button color on sign-up rates.


### Hypotheses for the A/B Test

**Null Hypothesis (H0):** There is no statistically significant difference in the sign-up rates between the control group (old button color) and the test group (new button color). Any observed difference is due to random chance.

$H_0: P_{\text{control}} = P_{\text{test}}$

**Alternative Hypothesis (H1):** There is a statistically significant difference in the sign-up rates between the control group (old button color) and the test group (new button color). Specifically, the new button color has an impact (either positive or negative) on the sign-up rate.

$H_1: P_{\text{control}} \neq P_{\text{test}}$

## Calculate Conversion Rates and Uplift

### Subtask:
Calculate the conversion rates for both the control and test groups. Then, compute the uplift (percentage change) in conversion rate between the test group and the control group.


**Reasoning**:
To calculate the conversion rates and uplift, I will first sum the 'purchase' and 'impressions' columns for both the control and test groups. Then, I will compute the conversion rate for each group and finally calculate the percentage uplift, printing all these results as requested.



````python
print('--- Calculating Conversion Rates and Uplift ---')

# 1. Calculate total purchases and impressions for the control group
total_purchases_control = control_df['purchase'].sum()
total_impressions_control = control_df['impressions'].sum()

# 2. Calculate total purchases and impressions for the test group
total_purchases_test = test_df['purchase'].sum()
total_impressions_test = test_df['impressions'].sum()

# 3. Compute conversion rate for the control group
control_conversion_rate = (total_purchases_control / total_impressions_control) * 100

# 4. Compute conversion rate for the test group
test_conversion_rate = (total_purchases_test / total_impressions_test) * 100

# 5. Calculate the uplift (percentage change) in conversion rate
uplift_percentage = ((test_conversion_rate - control_conversion_rate) / control_conversion_rate) * 100

# 6. Print the results
print(f"\nControl Group Conversion Rate: {control_conversion_rate:.4f}%")
print(f"Test Group Conversion Rate: {test_conversion_rate:.4f}%")
print(f"Uplift in Conversion Rate: {uplift_percentage:.4f}%")
````

## Perform Significance Test

### Subtask:
Conduct a statistical significance test (e.g., Chi-squared test) to compare the conversion rates of the control and test groups using a significance level (Î±) of 0.05.


**Reasoning**:
To perform the Chi-squared test, I will import the necessary function, use the previously calculated total purchases and impressions to construct a contingency table, and then apply the chi2_contingency function to determine the p-value and statistical significance.



````python
from scipy.stats import chi2_contingency

# Define the significance level
alpha = 0.05

# Calculate non-purchases for each group
total_non_purchases_control = total_impressions_control - total_purchases_control
total_non_purchases_test = total_impressions_test - total_purchases_test

# Create a 2x2 contingency table
# Rows: Control, Test
# Columns: Purchases, Non-Purchases
contingency_table = [
    [total_purchases_control, total_non_purchases_control],
    [total_purchases_test, total_non_purchases_test]
]

print("--- Performing Chi-squared Test ---")
print("Contingency Table (Purchases vs. Non-Purchases):")
print(contingency_table)

# Perform Chi-squared test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"\nChi-squared Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")

# Compare p-value to alpha and state conclusion
if p_value < alpha:
    print(f"\nSince the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("There is a statistically significant difference in conversion rates between the control and test groups.")
else:
    print(f"\nSince the P-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("There is no statistically significant difference in conversion rates between the control and test groups.")
````

## Calculate Confidence Intervals

### Subtask:
Calculate the confidence intervals for the conversion rates of both groups.


**Reasoning**:
To calculate the confidence intervals for the conversion rates, I will use the `proportion_confint` function from the `statsmodels` library, as specified in the instructions. I'll use the previously calculated total purchases and impressions for each group and a defined confidence level of 95%.



````python
from statsmodels.stats.proportion import proportion_confint

# Define the confidence level
confidence_level = 0.95
alpha_ci = 1 - confidence_level # alpha for proportion_confint is 1 - confidence_level

print("--- Calculating Confidence Intervals ---")

# Calculate confidence interval for the control group
# It's important to cast the number of purchases to an integer, as proportion_confint expects integer counts
control_purchases_int = int(round(total_purchases_control))
control_impressions_int = int(round(total_impressions_control))

lower_bound_control, upper_bound_control = proportion_confint(
    count=control_purchases_int,
    nobs=control_impressions_int,
    alpha=alpha_ci,
    method='wilson' # Wilson score interval is often preferred for A/B testing
)

# Calculate confidence interval for the test group
test_purchases_int = int(round(total_purchases_test))
test_impressions_int = int(round(total_impressions_test))

lower_bound_test, upper_bound_test = proportion_confint(
    count=test_purchases_int,
    nobs=test_impressions_int,
    alpha=alpha_ci,
    method='wilson'
)

# Print the confidence intervals as percentages
print(f"\nControl Group {confidence_level*100:.0f}% Confidence Interval for Conversion Rate: "
      f"({lower_bound_control*100:.4f}%, {upper_bound_control*100:.4f}%)")
print(f"Test Group {confidence_level*100:.0f}% Confidence Interval for Conversion Rate: "
      f"({lower_bound_test*100:.4f}%, {upper_bound_test*100:.4f}%)")
````

## Visualize Conversion Rates

### Subtask:
Create a bar chart or similar visualization comparing the conversion rates of the control and test groups, including confidence intervals or error bars if appropriate, to clearly illustrate the results. Ensure proper labels and legends.


**Reasoning**:
To visualize the conversion rates and their confidence intervals, I will use `matplotlib.pyplot` to create a bar chart as instructed. I will extract the necessary data (conversion rates and confidence interval bounds) from the previously computed variables, calculate the error for the error bars, and then plot and annotate the results.



````python
import matplotlib.pyplot as plt

print("--- Visualizing Conversion Rates with Confidence Intervals ---")

# 1. Create labels for the x-axis
labels = ['Control Group', 'Test Group']

# 2. Create conversion rates for the y-axis (as percentages)
conversion_rates = [control_conversion_rate, test_conversion_rate]

# 3. Calculate the error (half-width of the confidence interval) for both groups
error_control = (upper_bound_control - lower_bound_control) / 2 * 100
error_test = (upper_bound_test - lower_bound_test) / 2 * 100
errors = [error_control, error_test]

# 4. Create a bar chart comparing the conversion rates
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(labels, conversion_rates, yerr=errors, capsize=5, color=['skyblue', 'lightcoral'])

# 5. Add appropriate labels to the x-axis and y-axis
ax.set_xlabel('Group', fontsize=12)
ax.set_ylabel('Conversion Rate (%)', fontsize=12)

# 6. Add a title to the plot
ax.set_title('Conversion Rates with 95% Confidence Intervals', fontsize=14)

# 7. Add text annotations to each bar
for i, bar in enumerate(bars):
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + errors[i] + 0.02, # Position above error bar
            f'{yval:.4f}%\nCI: ({conversion_rates[i] - errors[i]:.4f}%, {conversion_rates[i] + errors[i]:.4f}%)',
            ha='center', va='bottom', fontsize=9, color='black')

ax.set_ylim(bottom=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# 8. Display the plot
plt.show()
````

### How We Conducted the Test: Our Methodology

To determine the effect of the new button color, we split our audience into two groups:

*   **Control Group:** Users in this group saw our existing button color. This served as our baseline.
*   **Test Group:** Users in this group were exposed to the new button color.

Both groups were shown the respective button colors for the same duration, and we meticulously tracked key metrics, primarily focusing on **conversion rate** â the percentage of users who completed the sign-up process after viewing the button. By comparing these two groups, we could isolate the impact of the new button color, minimizing external factors.

Since this p-value is significantly lower than our chosen significance level of 0.05, we can confidently reject the idea that the difference is random. This means the new button color had a **statistically significant positive impact** on sign-up rates.

*   **Control Group 95% Confidence Interval:** (0.4698%, 0.4847%)
*   **Test Group 95% Confidence Interval:** (0.6880%, 0.7098%)

These intervals tell us that if we were to repeat this experiment many times, 95% of the time, the true conversion rate for the control group would fall between 0.4698% and 0.4847%, and for the test group, between 0.6880% and 0.7098%.
